```python
import pandas as pd
```

```python
# Retrieve the dataset
path = "household_power_consumption.txt"

# Load the dataset into a dataframe
df = pd.read_csv(path, delimiter = ";", low_memory = False)

df
```

```python
# Print the length of the dataset
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
# Look up the numer of missing values
print(df.isnull().sum())
```

```python
# Drop rows with missing values
df = df.dropna()

# Fill remaining missing values with the mean of the column
# df[['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']] = df[['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].fillna(df[['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean())
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
# Convert the date column to datetime object
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

# Split the DataFrame into three datasets based on the year
df_2007 = df[df['Date'].dt.year == 2007]
df_2008 = df[df['Date'].dt.year == 2008]
df_2009 = df[df['Date'].dt.year == 2009]

# Save each dataset to a separate file
df_2007.to_csv('dataset_2007.csv', index=False)
df_2008.to_csv('dataset_2008.csv', index=False)
df_2009.to_csv('dataset_2009.csv', index=False)
```
